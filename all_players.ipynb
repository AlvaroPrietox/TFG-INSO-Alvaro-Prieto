# All players

En este archivo lo que hacemos es juntar todos los datos de los jugadores de todas las temporadas y años en uno solo

Para ello hemos añadido dos columnas:

- *League*: liga a la que pertenece
- *Season*: temporada a la que pertenece

```python
import pandas as pd
from pathlib import Path
```

```python
BASE_PATH = Path("datos")
```

```python
dfs = []

for league_path in BASE_PATH.iterdir():
    if not league_path.is_dir():
        continue

    league = league_path.name  # bundesliga, laliga, etc.

    for season_path in league_path.iterdir():
        if not season_path.is_dir():
            continue

        season = season_path.name  # 2020-2021

        players_file = season_path / "players.csv"
        if not players_file.exists():
            continue

        df = pd.read_csv(players_file)

        # Añadimos variables estructurales
        df["league"] = league
        df["season"] = season

        dfs.append(df)
```

```python
players_all = pd.concat(dfs, ignore_index=True)
```

```python
import os

# Ruta base
base_path = 'datos/La-Liga/2025-2026'

# Lista para almacenar todos los dataframes
all_data = []

# Recorremos cada carpeta de jornada
for jornada in os.listdir(base_path):
    jornada_path = os.path.join(base_path, jornada)

    # Verificamos que sea una carpeta
    if os.path.isdir(jornada_path):

        # Recorremos los archivos dentro de la jornada
        for file in os.listdir(jornada_path):
            if file.endswith('.csv'):
                file_path = os.path.join(jornada_path, file)

                # Leemos el CSV
                df = pd.read_csv(file_path)

                # Añadimos la columna jornada con el nombre de la carpeta
                df['jornada'] = jornada

                # Guardamos el dataframe
                all_data.append(df)

# Concatenamos todos los dataframes
players_laliga = pd.concat(all_data, ignore_index=True)

# Guardamos el dataset final
output_path = os.path.join(base_path, 'players_laliga.csv')
players_laliga.to_csv(output_path, index=False)

print(f'Dataset creado correctamente en: {output_path}')
print(players_laliga.head())
```

```python
base_path = 'datos'

matches_data = []

# Recorremos cada liga
for liga in os.listdir(base_path):
    liga_path = os.path.join(base_path, liga)

    if os.path.isdir(liga_path):

        # Recorremos cada temporada
        for temporada in os.listdir(liga_path):
            temporada_path = os.path.join(liga_path, temporada)

            if os.path.isdir(temporada_path):

                # Buscamos CSVs tipo season-2021.csv
                for file in os.listdir(temporada_path):
                    if file.endswith('.csv') and file.startswith('season'):
                        file_path = os.path.join(temporada_path, file)

                        try:
                            df = pd.read_csv(file_path)

                            # Añadimos columnas de contexto
                            df['liga'] = liga
                            df['temporada'] = temporada

                            matches_data.append(df)

                        except Exception as e:
                            print(f'Error leyendo {file_path}: {e}')

# Unimos todos los partidos
matches_dataset = pd.concat(matches_data, ignore_index=True)

# Guardamos dataset final
output_matches = os.path.join(base_path, 'matches_dataset.csv')
matches_dataset.to_csv(output_matches, index=False)

print(f'Dataset de partidos creado en: {output_matches}')
print(matches_dataset.head())
```
