```python
import pandas as pd
```

```python
players_all = pd.read_csv("all_players.csv")
```

```python
players_all.head()
```

```python
stats_cols = [
    "goals", "xG", "assists", "xA", "shots",
    "key_passes", "yellow_cards", "red_cards",
    "npg", "npxG", "xGChain", "xGBuildup"
]
```

```python
MIN_MINUTES = 900

players_all = players_all[players_all["time"] >= MIN_MINUTES].copy()
```

```python
for col in stats_cols:
    players_all[f"{col}_90"] = players_all[col] / players_all["time"] * 90
```

```python
players_all.drop(columns=stats_cols, inplace=True)
```

```python
players_all.to_csv("all_players_normalized.csv", index=False)
```
